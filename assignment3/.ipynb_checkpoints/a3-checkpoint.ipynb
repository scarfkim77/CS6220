```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_data = iris['data']
iris_target = iris['target']
iris_names = iris['feature_names']
```

Part 1 - PCA

1. Apply PCA projection of the features of IRIS dataset in 2 dimensions

```python
from sklearn.decomposition import PCA
# 2 dimensions
pca = PCA(n_components=2)
pca.fit(iris_data)
```

2. Show how much variance ratio is explained by the reduced dimension.

```python
print(pca.explained_variance_ratio_)
```

Part 2 - Another Reduction Method

1. Apply another reduction method of your choice on the features of IRIS dataset in 2
dimensions.

    I'd like to use Linear Discriminant Analysis(LDA) as another reduction method.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
# 2 dimensions
lda = LDA(n_components=2)
lda_reduction = lda.fit_transform(iris.data, iris.target)
```

2. Show scatter plot of the reduced dimension. Use separate color for each class of the
data.

```python
import matplotlib.pyplot as plt
# set size
plt.figure(figsize=(13 ,3))
# plot the 3 classes separately in different colors
plt.scatter(lda_reduction[iris.target == 0, 0], lda_reduction[iris.target == 0, 1], color='r')
plt.scatter(lda_reduction[iris.target == 1, 0], lda_reduction[iris.target == 1, 1], color='g')
plt.scatter(lda_reduction[iris.target == 2, 0], lda_reduction[iris.target == 2, 1], color='b')
plt.title('LDA of Iris', fontsize=13, pad=13)
plt.show()
```

3. Show how much variance ratio is explained by the reduced dimension.

```python
print(lda.explained_variance_ratio_)
```

Part 3 Conceptual Question

1. Compare the variance ratio explained by the 2-dimensions of the methods you have
used. Which is better?

    The variance ratio of PCA is: [0.92461872 0.05306648]
    The variance ratio of LDA is: [0.9912126 0.0087874]
    We can see that LDA performs better.

2. Compare the scatter plot of the two methods after reduction. Which is a better method
for separating the different classes of data?

    Below is the scatter plot of the two methods. 
    We can see LDA performs better for separating the different classes of data.
    The boundaries between different classes are clearer than PCA.

```python
pca = PCA(n_components=2)
pca_reduction = pca.fit_transform(iris.data)
# set size
plt.figure(figsize=(13 ,3))
plt.subplot(1, 2, 1)
# plot the 3 classes of PCA separately in different colors
plt.scatter(pca_reduction[iris.target == 0, 0], pca_reduction[iris.target == 0, 1], color='r')
plt.scatter(pca_reduction[iris.target == 1, 0], pca_reduction[iris.target == 1, 1], color='g')
plt.scatter(pca_reduction[iris.target == 2, 0], pca_reduction[iris.target == 2, 1], color='b')
plt.title('PCA of Iris', fontsize=13, pad=13)

plt.subplot(1, 2, 2)
# plot the 3 classes of LDA separately in different colors
plt.scatter(lda_reduction[iris.target == 0, 0], lda_reduction[iris.target == 0, 1], color='r')
plt.scatter(lda_reduction[iris.target == 1, 0], lda_reduction[iris.target == 1, 1], color='g')
plt.scatter(lda_reduction[iris.target == 2, 0], lda_reduction[iris.target == 2, 1], color='b')
plt.title('LDA of Iris', fontsize=13, pad=13)
plt.show()
```

3. What is the primary difference between the two methods? Which method works better
in this case and why?

        The major difference is that PCA is unsupervised, ignoring class labels, but LDA is a supervised dimensionality reduction method, taking class labels into consideration.
        PCA is trying to find a set of uncorrelated components of maximum variance in a dataset, focues on capturing the directions of maximal variance.
        On the contrary, LDA focues on finding the 1 dimension that maximizes class separability between the groups.
        In our case, LDA is better, because we need to separate cluster, thus we have to know clusters before. LDA takes class labels into account, and aims at separating data of different categories as much as possible.
